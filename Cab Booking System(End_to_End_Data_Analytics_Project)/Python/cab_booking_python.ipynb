# 10 High-Value Analytical Questions for a Cab Booking System

```python
# 1. What is the overall cancellation rate?
cancellation_rate=(df["ride_status"]=="cancelled").mean()*100
cancellation_rate
```

```python
# 2. Which city generates the highest number of rides?
data=df["city"].value_counts()
data
```

```python
# 3. Which city generates the highest revenue?
d1=df.groupby("city")["fare"].sum().sort_values(ascending=False)
d1
```

```python
# 4. What is the average fare per ride by vehicle type?
d2=df.groupby("vehicle_type")["fare"].mean().sort_values(ascending=False)
d2
```

```python
# 5. Which payment method is used most frequently?
d3=df["payment_method"].value_counts()
d3
```

```python
# 6. What time of day has the highest ride demand?

df["hour"]=df["pickup_time"].dt.hour
d4=df["hour"].value_counts().sort_values(ascending=False)
d4
```

```python
# 7. Which drivers complete the most rides?
d5=df[df["ride_status"]=="completed"].groupby("driver_name")["ride_id"].count().sort_values(ascending=False)
d5
```

```python
# 8. What is the average ride duration?
df["ride_duration"]=df["drop_time"]-df["pickup_time"]
avg=df["ride_duration"].mean()
avg
```

```python
# 9. Which routes are most popular?
df.groupby(["pickup_location","drop_location"])["ride_id"].count().sort_values(ascending=False)
```

```python
# 10. What percentage of payments fail?

(df["payment_status"]=="failed").mean()*100
```
